```python
import os

from skimage import io
from matplotlib import pyplot as plt

import intelligent_placer_lib as ipl
from intelligent_placer_lib.modules import detector
```

```python
file_name = "./data/items_sets/1.jpg"
polygons_path = "./data/polygon_samples"

plt.figure(figsize=(7,7))
img = io.imread(file_name)

plt.imshow(img)
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i, imobj in enumerate(detector.image_object_range(img, (256, 256), 800)):
    ax[i].imshow(imobj)
plt.show()
```

```python
for pol_name in os.listdir(polygons_path):
    polygon = ipl.load_polygon(os.path.join(polygons_path, pol_name))
    print(f"{pol_name} {ipl.check_image(img, polygon)}")
```
